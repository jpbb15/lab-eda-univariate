Part1: Understanding Product Categories

```python
#Generate a frequency table for the product category
products_uk_df['category'].value_counts()
```

```python
#Top 5 most listed products categories
top5_products_categories = products_uk_df['category'].value_counts().head(5)
top5_products_categories
```

```python
#Display the distribution of products across different categories using a bar chart
import matplotlib.pyplot as plt
top5_products_categories.plot(kind='bar')
plt.title('Top 5 most listed products categories')
plt.xlabel('Categories')
plt.ylabel('Frequency')
plt.show()
```

```python
#For a subset of top categories, visualize their proportions using a pie chart
top5_products_categories.plot(kind='pie', autopct='%1.1f%%')
plt.title('Top 5 most listed products categories')
plt.show()
```

Part 2: Delving into Product Pricing

```python
#Calculate the mean, median, and mode for the price of products
print("The mean is:", products_uk_df['price'].mean().round(2),"€")
print("The median is:", products_uk_df['price'].median())
print("The mode is:", products_uk_df['price'].mode())
```

```python
#What's the average price point of products listed? How does this compare with the most common price point (mode)?
products_uk_df['price'].describe()
#The mean is 10x higher than the mode which means that the price distribution is right-skewed
```

```python
#Determine the variance, standard deviation, range, and interquartile range for product prices
print("The variance is:", products_uk_df['price'].var())
print("The standard deviation is:", products_uk_df['price'].std())
print("The range is:", products_uk_df['price'].max() - products_uk_df['price'].min())
print("The interquartile range is:", products_uk_df['price'].quantile(0.75) - products_uk_df['price'].quantile(0.25))
```

```python
#There are a significant spread and variability in the prices of products listed
```

```python
#Plot a histogram to visualize the distribution of product prices.
products_uk_df['price'].plot(kind='hist', bins=20)
plt.title('Distribution of product prices')
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.show()
```

```python
#Use a box plot to showcase the spread and potential outliers in product pricing
products_uk_df['price'].plot(kind='box')
plt.title('Distribution of product prices')
plt.ylabel('Price')
plt.show()
```

Part 3: Unpacking Product Ratings

```python
#Calculate the mean, median, and mode for the rating of products.
print("The mean is:", products_uk_df['stars'].mean())
print("The median is:", products_uk_df['stars'].median())
print("The mode is:", products_uk_df['stars'].mode())
```

```python
#Costumers are not generally satisfied with the products as the mean rating is below 3
```

```python
#Determine the variance, standard deviation, and interquartile range for product rating.
print("The variance is:", products_uk_df['stars'].var())
print("The standard deviation is:", products_uk_df['stars'].std())
print("The interquartile range is:", products_uk_df['stars'].quantile(0.75) - products_uk_df['stars'].quantile(0.25))
```

```python
#Calculate the skewness and kurtosis for the rating column
print("The skewness is:", products_uk_df['stars'].skew())
print("The kurtosis is:", products_uk_df['stars'].kurt())
```

```python
#The ratngs are negatively skewed and have a platykurtic distribution
```

```python
#Plot a histogram to visualize the distribution of product ratings
products_uk_df['stars'].plot(kind='hist', bins=20)
plt.title('Distribution of product ratings')
plt.xlabel('Rating')
plt.ylabel('Frequency')
plt.show()
```
